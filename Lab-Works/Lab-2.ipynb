# Working with Dictionary and DataFrame

## Part I: Working with Dictionary
- A data structure used to store data in **key-value pairs**

```python
student = {
  "name" : "Hari",
  "age" : 20,
  "faculty": "BCE"
}
```

Tasks: Perform following operations

## Create Dictionary -> d = {}

````python
student = {
  "name" : "Hari",
  "age" : 20,
  "faculty": "BCE"
}
````

## Access Value -> d[key]

````python
print(student["name"])
````

## Using get() -> d.get(key)

````python
print(student.get("age"))
````

## Adding Item -> d[key] = value

````python
student['group'] = 'B'
print(student)
````

## Update Vlue -> d[key] = new value

````python
student['name'] = 'Shyam'
print(student)
````

## Remove Item using del -> del d[key]

````python
del student['group']
print(student)
````

## Remove using pop -> d.pop(key)

````python
student.pop("name")
````

````python
print(student)
````

## Remove Last Item -> d.popitem()

````python
student.popitem()
````

````python
print(student)
````

## Clear dictionary -> d.clear()

````python
student.clear()
print(student)
````

## Copy Dictionary -> d.copy()

````python
student = {
  "name" : "Hari",
  "age" : 20,
  "faculty": "BCE"
}

student_copy = student.copy()
print(student_copy)
````

## Get all keys -> d.keys()

````python
print(student.keys())
````

## Get all values -> d.values()

````python
print(student.values())
````

## Get all items -> d.items()

````python
print(student.items())
````

## Check Key Exits -> key in d

````python
'name' in student
````

## Get Length -> len(d)

````python
len(student)
````

## Loop Through Keys -> for k in d:

````python
for keys in student:
  print(keys)
  print(student[keys])
````

## Loop Through Values -> for v in d.values():

````python
for values in student.values():
  print(values)
````

Loop Through Items -> for k, v in d.items():

````python
for keys, values in student.items():
  print(keys, values)
````

## Part II: Working with Pandas DataFrame

````python
import pandas as pd

data = {
    'Name': ['Ram', 'Shyam', 'Hari', 'Sita', 'Gita', 'Pdf'],
    'Age': [24, 27, 22, 32, 20, 22],
    'City': ['Kathmandu', 'Bhaktapur', 'Bajhang', 'Lalitpur', 'Biratnagar', 'idk'],
    'Marks': [14, 77, 62, 32, 80, 92]
}
````

## Creating a DataFrame -> df.= pd.DataFrame()

````python
df = pd.DataFrame(data)
````

## Displaying the first 5 rows of the DataFrame -> df.head()

````python
df.to_csv('my_dataframe.csv', index=False)
print('DataFrame saved to my_dataframe.csv')
````

````python
df.tail()
````

## Concatenate two DataFrames vertically -> pd.concat([df1, df2])

````python
data1 = {
    'Name': ['Ram', 'Shyam', 'Hari'],
    'Age': [24, 27, 22],
    'City': ['Kathmandu', 'Bhaktapur', 'Lalitpur']
}
data2 = {
    'Name': ['Sita', 'Gita', 'Rita'],
    'Age': [21, 21, 22],
    'City': ['Bajhang', 'Bajura', 'Doti']
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print('<-----------df1----------->')
print(df1)
print('<-----------df2----------->')
print(df2)

concatenated_df = pd.concat([df1, df2])
print('<------------concatenated df---------->')
print(concatenated_df)
````

## Write the DataFrame into an Excel file -> df.to_excel('file.xslx')
